# Selenium

##### 自动化测试工具，支持多种浏览器。
##### 爬虫中重要用来解决Javascript渲染的问题

### 基本使用

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.wait import WebDriverWait

# 声明浏览器对象
browser = webdriver.Chrome()
try:
    #访问页面
    browser.get('https://www.baidu.com')
    #寻找相关元素 find_element
    input = browser.find_element_by_id('kw')
    input.send_keys('Python')
    input.send_keys(Keys.ENTER)
    wait = WebDriverWait(browser,10)
    #等待content_left元素加载
    wait.until(EC.presence_of_element_located((By.ID,'content_left')))
    print(browser.current_url)
    print(browser.get_cookies())
    print(browser.page_source)
finally:
    browser.close()
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By

browser = webdriver.Chrome()
browser.get('https://www.taobao.com')
input_first = browser.find_element(By.ID,'q')
print(input_first)
browser.close()
```

## 多个元素

#### find_elements_by_name
#### find_elements_by_xpath
#### find_elements_by_link-text
#### find_elements_by_partial_link_text
#### find_elements_by_tag_name
#### find_elements_by_class_name
#### find_elements_by_css_selector

```python
from selenium import webdriver
from selenium.webdriver.common.by import By

browser = webdriver.Chrome()
browser.get('https://www.taobao.com')
list = browser.find_elements(By.CSS_SELECTOR,'.service-bd li')
print(list)
browser.close()
```

## 元素交互操作

```python
from selenium import webdriver
import time

browser = webdriver.Chrome()
browser.get('https://www.bilibili.com')
input = browser.find_element_by_class_name('search-keyword')
input.send_keys('iPone')
time.sleep(4)
input.clear()
input.send_keys('iPad')
input.send_keys(Keys.ENTER)
button = browser.find_element_by_class_name('search-submit')
button.click()
```
